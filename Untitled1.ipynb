```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# 1. Cargar el dataset MNIST (Dígitos escritos a mano)
# Son 60,000 imágenes de entrenamiento y 10,000 de prueba
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# 2. Normalizar los datos (De 0-255 a 0-1)
x_train, x_test = x_train / 255.0, x_test / 255.0

print("Datos cargados. Dimensiones de una imagen:", x_train[0].shape)
```

```python
# Visualizar qué se está analizando
plt.figure(figsize=(10,5))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(f"Etiqueta: {y_train[i]}")
    plt.axis('off')
plt.show()
```

```python
# Agregamos una dimensión extra a los datos para que funcionen con Convoluciones
# Las CNN esperan (Ancho, Alto, Canales de color) -> (28, 28, 1)
x_train_cnn = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test_cnn = x_test.reshape(x_test.shape[0], 28, 28, 1)

modelo = tf.keras.models.Sequential([
    # 1. Capa Convolucional: 32 filtros de 3x3.
    # Esto es el equivalente a implementar SIFT en SEM.
    # Busca bordes y líneas pequeñas.
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),

    # 2. MaxPooling: Reduce la imagen quedándose con la información más fuerte.
    # Hace que el modelo sea tolerante a pequeñas variaciones de posición.
    tf.keras.layers.MaxPooling2D((2, 2)),

    # 3. Otra Capa Convolucional: Busca patrones más complejos (curvas del '3' o bucles del '8')
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),

    # 4. Aplanar y Clasificar (lo que ya se tenía en el anterior código)
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

modelo.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("✅ Modelo Convolucional (CNN) construido. Este es mucho más robusto.")

# Entrenamos con los datos reformateados
# Subimos a 5 o 10 épocas porque este modelo aprende más detalles
print("Entrenando CNN...")
modelo.fit(x_train_cnn, y_train, epochs=5)
print("¡Listo!")
```

```python
# Evaluar con datos que la red NUNCA ha visto (Test set)
test_loss, test_acc = modelo.evaluate(x_test,  y_test, verbose=2)
print(f'\nPrecisión en el set de prueba: {test_acc*100:.2f}%')
```

```python
#@title Panel de prueba
from IPython.display import HTML, Image
from google.colab.output import eval_js
import google.colab
from base64 import b64decode
import numpy as np
from PIL import Image as PILImage
import io
import cv2

canvas_html = """
<canvas width=280 height=280 style="border:2px solid #000000; background-color: black;"></canvas>
<br>
<button onclick="clear_canvas()">Borrar</button>
<button onclick="finish()">¡Predecir!</button>
<script>
var canvas = document.querySelector('canvas');
var ctx = canvas.getContext('2d');
ctx.lineWidth = 25; // Pincel grueso para que se parezca a MNIST
ctx.strokeStyle = 'white';
var mouse = {x: 0, y: 0};

canvas.addEventListener('mousemove', function(e) {
  mouse.x = e.pageX - this.offsetLeft;
  mouse.y = e.pageY - this.offsetTop;
}, false);

canvas.addEventListener('mousedown', function(e) {
    ctx.beginPath();
    ctx.moveTo(mouse.x, mouse.y);
    canvas.addEventListener('mousemove', onPaint, false);
}, false);

canvas.addEventListener('mouseup', function() {
    canvas.removeEventListener('mousemove', onPaint, false);
}, false);

var onPaint = function() {
    ctx.lineTo(mouse.x, mouse.y);
    ctx.stroke();
};

function clear_canvas() {
    ctx.clearRect(0, 0, canvas.width, canvas.height);
}

function finish() {
    var data = canvas.toDataURL('image/png');
    google.colab.kernel.invokeFunction('notebook.process_image', [data], {});
}
</script>
"""

def process_image(data):
    # 1. Decodificar la imagen del navegador
    binary = b64decode(data.split(',')[1])
    img = PILImage.open(io.BytesIO(binary)).convert('L')

    # 2. Redimensionar a 28x28 (Tamaño MNIST)
    img = img.resize((28, 28))
    img_array = np.array(img)

    # 3. Normalizar (0-255 -> 0-1)
    img_array = img_array / 255.0

    # 4. Mostrar lo que la red "ve" (Para depuración visual)
    import matplotlib.pyplot as plt
    plt.imshow(img_array, cmap='gray')
    plt.title("Lo que ve la red (28x28)")
    plt.axis('off')
    plt.show()

    # 5. Preprocesar para la CNN
    # La CNN espera: (Batch, Alto, Ancho, Canales) -> (1, 28, 28, 1)
    img_array = img_array.reshape(1, 28, 28, 1)

    # 6. Predicción usando 'modelo' (modelo CNN)
    prediction = modelo.predict(img_array)
    predicted_digit = np.argmax(prediction)
    confianza = np.max(prediction) * 100

    print(f"\n>>> PREDICCIÓN: {predicted_digit} <<<")
    print(f"Confianza: {confianza:.2f}%")

    # Mostrar probabilidades completas (opcional, para ver dudas)
    # print("Probabilidades:", prediction)

google.colab.output.register_callback('notebook.process_image', process_image)
display(HTML(canvas_html))
```
